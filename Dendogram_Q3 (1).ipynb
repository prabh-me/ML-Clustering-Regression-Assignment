The code is straightforward. The upper diagonal distance matrix has been converted to a condensed numpy array and dendogram has been created after performing hierarchical clustering.

For single link clustering, any two high distance values can be modified and the dendogram would remain the same.

For complete link clustering, any two low distance values can be modified and the dendogram would remain the same.

```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
import numpy as np
```

```python
ytdist = np.array([0.12, 0.51, 0.84, 0.28, 0.34, 0.25, 0.16, 0.77, 0.61, 0.14, 0.70, 0.93, 0.45, 0.20, 0.67])
Z = hierarchy.linkage(ytdist, 'single')
plt.figure()
dn = hierarchy.dendrogram(Z)

plt.show()
```

```python
Z = hierarchy.linkage(ytdist, 'complete')
plt.figure()
dn = hierarchy.dendrogram(Z)


hierarchy.set_link_color_palette(['m', 'c', 'y', 'k'])

plt.show()
```
